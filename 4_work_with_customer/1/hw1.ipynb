### Курс "Машинное обучение"

Мы хотим, чтобы курс "Машинное обучение" купили как можно больше людей. При проведении аналитики будем использовать следующие метрики:
1. качество материала
2. количество слушателей
3. компетентность преподавателей

В поиске инсайтов нам может помочь фидбек слушателей (проходящих обучение в настоящий момент или уже прослушавших курс)

Команда может состоять из владельца продукта (чтобы взаимодействовать с ним по поводу организации учебной деятельности и узнавать все подводные камни), маркетолога (чтобы помогал организовать рекламную кампанию), BI-аналитика (чтобы проводил опросы слушателей для получения адекватного фидбека) и аналитика.

Тогда процесс аналитики может выглядеть следующим образом:
- изучаем, насколько качественно организован учебный процесс, спланированы практические и лекционные занятия, какое количество часов отведено на изучение различных тем (в том числе, на углубленное изучение и домашние задания);
- получаем фидбек от слушателей, анализируем результаты, вносим корректировки в учебную программу, нанимаем новых специалистов, вносим изменения в текущий преподавательских состав;
- разворачиваем рекламную кампанию (таргетированная реклама, баннеры, объявления, семинары и открытые занятия для студентов старших курсов в вузах и т.д.).
